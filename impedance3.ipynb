# LCR in series

```python
import lcr
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
L=1    # inductance in H
C=1e-6 #capacitance in F
V=100  #p.d. in V
```

```python
f0=lcr.fo(L,C) # resonating frquency
f0
```

```python
freq = np.arange(10,320,10)
R = np.arange(100,700,100)
plt.figure(figsize = [10,6])
c=['y','b','c','g','k','m'] # color
for i in range(len(R)):
    imp=lcr.impedance(L,C,R[i],freq) #impedance
    cur=V/imp           #current
    Q=lcr.Q(L,C,R[i])   # Q faqctor 
    plt.plot(freq,cur, color=c[i],label = "R="+str(R[i])+r"$\Omega$ , Q="+str(Q))
    
plt.xlabel('Frequency(Hz)',fontsize =14)
plt.ylabel('Current(A)',fontsize =14)
plt.legend()
plt.title('LCR circuit',fontsize =18,color='r')
plt.show()
```
